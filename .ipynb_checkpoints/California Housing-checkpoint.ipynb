### Standard Scaler

```python
#all features all calculated to different different units. for GD to converge faster (come to global minima) we'll bring all data to same scale.
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

### Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
#print coeff
print(reg.coef_) #for every independent feature we have coefficient
```

```python
#on which parameters the model has been trained
reg.get_params()
```

```python
##prediction with test data
reg_pred=reg.predict(X_test)
reg_pred
```

### Assumptions

```python
## plot scatter plot for the prediction
plt.scatter(y_test,reg_pred) #if plotting is linear means model has done well
```

```python
#errors
residuals=y_test-reg_pred
```

```python
residuals
```

```python
#plotting residuals
sns.displot(residuals) # this should be normally distributed. #2nd assumption 
```

```python
#Scatter plot with respect to prediction and resiudals
#uniform distribution between test data and residuals should be there.
plt.scatter(reg_pred,residuals)
```

### Scoring the model

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_absolute_error(y_test,reg_pred)))
```

```python
#r2 and Adjusted r2
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
